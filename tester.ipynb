```python
def torr_sizing(t1,t2,cp,mfr):
    x = PrettyTable()
    x.field_names=["Diameter","Wall Temp","Length"]
    reactor_diameter = np.arange(0.0254,1.0,0.0254)
    wall_temp = np.arange(t2+200,t2+300,10.0)
    print(wall_temp)
    results = np.zeros(shape=(len(reactor_diameter),len(wall_temp)))
    print(len(reactor_diameter),len(wall_temp))

    for i in range(0,len(reactor_diameter)):
        for j in range(0,len(wall_temp)):
            T = maths.kelvin(wt)
            l = ((32*u_g*(2-oslash_a))/(5*P*oslash_a))*np.sqrt((R_g*T)/(2*np.pi*M))
            laod = ((2*l)+(2*a))/d
            h_wp = ((4*k_g)/d)*(((1+laod)*np.log(1+(1/laod)))-1)
            h_ws = (oslash_w*h_wp)+((1-oslash_w)*((2*k_g/d)/(np.sqrt(2)+laod)))
            results[i,j] = -np.log((wt-t2)/(wt-t1))*((mfr*cp)/(h_ws*np.pi*dia))
            x.add_row([reactor_diameter[i],wall_temp[j],results[i,j]])
    print(x)
        
    fig = go.Figure()
    for i in range(0,len(wall_temp)):
        fig.add_trace(go.Scatter(x=reactor_diameter,y=results[:,i],name=(str(round(wall_temp[i],2)))))
    fig.update_xaxes(title="Reactor Diameter (m)")
    fig.update_yaxes(title="Reactor Length (m)")
    fig.update_layout(
        showlegend=True,
        legend=dict(title="Wall Temperature (degC)"),
        title = "Minimum Reactor Wall Temperature Requirement at ",
        title_x = 0.5
    )
    return fig

temp1 = 180
temp2 = 300
cp_this = 2.6181
mfrate = 100.0
torr_sizing(temp1,temp2,cp_this,mfrate)
```
